```python
def delete_punct(inputStr):
    puncts = string.punctuation+"«»—–“”"
    inputStr = re.sub('['+puncts+']', '', inputStr)
    return inputStr.lower()


def get_list_files(path):
    return glob.glob(path)

def get_script_content(scirpt_files):
    text_list = []
    for filename in scirpt_files:
        try:
            with open(filename, 'r', encoding = "utf-8") as file_object: ## utf-8-sig: remove bom: \ufeff, ,encoding='utf-8'
                lines = file_object.readlines()
                text_list.append(lines[0].rstrip())
        except UnicodeError:
            with open(filename, 'r', encoding = "utf-16") as file_object: ## utf-8-sig: remove bom: \ufeff, ,encoding='utf-8'
                lines = file_object.readlines()
                text_list.append(lines[0].rstrip())
    return text_list

def get_target_files(target_filename_dir,training_files):
    text_list = {}
    for filename in training_files:
        i = filename.rindex('\\')+1
        clear_name = filename[i:len(filename)-4]
        textfile = target_filename_dir+clear_name+'.txt'
        print(textfile)
        try:
            with open(textfile, 'r', encoding = "utf-8") as file_object: ## utf-8-sig: remove bom: \ufeff, ,encoding='utf-8'
                lines = file_object.readlines()
                text_list.append(lines[0].rstrip())
        except UnicodeError:
            print("Error")
            with open(textfile, 'r', encoding = "utf-16") as file_object: ## utf-8-sig: remove bom: \ufeff, ,encoding='utf-8'
                lines = file_object.readlines()
                text_list[filename] = lines[0].rstrip()
    return text_list

def get_target_files_ru_ru_THU(target_filename_dir,training_files):
    script_lists = {}
    k = 0
    for filename in training_files:
        k += 1
        if k > 1000:
            break
        
        i = filename.rindex('\\')+1
        clear_name = filename[i:len(filename)-4].split('-')
        speakerID = clear_name[2]
        line_num = clear_name[3]

        # Find scirpt files by SpeakerID and the script by line number
        target_dir = target_filename_dir+'Speaker'+speakerID+'/'
        if int(line_num) <= 149:
            target_dir += "Session0.txt"
        else:
            target_dir += "Session1.txt"
        
        with open(target_dir, 'r') as f:
            lines = f.readlines()
            for line in lines:
                ln = line.strip().split('\t')[0]
                if ln == line_num:
                    script_lists[filename] = line.strip().split('\t')[1]

    return script_lists

def sparse_tuple_from(sequences, dtype=np.int32):
    indices = []
    values = []

    for n, seq in enumerate(sequences):
        indices.extend(zip([n]*len(seq), range(len(seq))))
        values.extend(seq)

    indices = np.asarray(indices, dtype=np.int64)
    values = np.asarray(values, dtype=dtype)
    shape = np.asarray([len(sequences), np.asarray(indices).max(0)[1]+1], dtype=np.int64)
    return indices, values, shape
```

```python
#kazakh_alphabetic_order
def KAO(ch, flag):
    alphabet = ["a","b","c","d","e","f","g","h","i","j","k","l","m","n","o","p","q","r","s","t","u","v","w","x","y","z","-"]
    if flag == 1:
        out = alphabet.index(ch)+1 #index
    elif flag == 0:
        out = alphabet[ch-1] if ch!=0 else " " #character
    return out

# Pre-Processing
def pre_processing(dic_audio_script):
    features = []
    labels = []
    seq_lengths = []
    originals = []
    max_len = 0
    for audio_file,audio_scirpt in dic_audio_script.items():
        # Process audio wav file    
        fs, audio = wav.read(audio_file)
        print('fs ',fs)
        inputs = mfcc(audio, samplerate=fs)
        train_inputs = (inputs - np.mean(inputs))/np.std(inputs)
        train_seq_len = [train_inputs.shape[0]]

        if max_len < train_inputs.shape[0]:
            max_len = train_inputs.shape[0]
    
        # Process audio scirpts to unicodes
        text = delete_punct(audio_scirpt)

        targets = text.replace(' ', '  ')
        targets = targets.split(' ')    
        targets = np.hstack([SPACE_TOKEN if x == '' else list(x) for x in targets])# change " " to <space>, set each character as a member of the list
    
        unicode_targets = np.asarray([SPACE_INDEX if x == SPACE_TOKEN else KAO(x,1) for x in targets])# set unicodes, subtract the FIRST_INDEX to start encode from the value of one 
        print('train_inputs ',train_inputs.shape)
        # Set these in their lists
        features.append(train_inputs)
        labels.append(unicode_targets)
        seq_lengths.extend(train_seq_len)
        originals.append(text)

    return features, np.array(labels), seq_lengths, originals

```

```python
features, label_array, seq_lengths, originals = pre_processing(training_targets)
```

```python
# next batch generator
def next_batch_gen(batch_counter):
    start = batch_counter * batch_size
    end = (batch_counter+1) * batch_size

    batchfeature = features[start:end]
    batch_label = label_array[start:end]
    batch_seq_lengths = seq_lengths[start:end]
    batch_seq_originals = originals[start:end]
    max_len = max(batch_seq_lengths)
    print(num_features)
    padded_features = np.zeros([batch_size, max_len, num_features], dtype=np.float32)
    # padding
    for i, feature in enumerate(batchfeature):
        pad_length = max_len - len(feature)
        padded_feature = feature.tolist() + [[PAD_VALUE] * num_features] * pad_length
        padded_features[i] = np.array(padded_feature);
  
    return padded_features, batch_label, batch_seq_lengths, batch_seq_originals
```

```python
label_seq_lengths = [el.size for el in label_array]

def batch_gen(features, label_array, seq_lengths, originals, batch_size=5):
    idx = np.random.randint(len(features), size=batch_size)
    label_seq_lengths = []
    batch_features = np.asarray(features)
    batch_features = batch_features[idx]
    batch_label = label_array[idx]
    
    max_len = max([el.shape[0] for el in batch_features])
    for i,label in enumerate(batch_label):
        pad_len = max(max_len-len(label),0)
        buff = np.pad(label,(0,pad_len),'constant', constant_values=(PAD_VALUE))
        batch_label[i]=buff
        label_seq_lengths.append(len(label))


    batch_label = Variable(torch.IntTensor(np.stack(batch_label,axis=0)), requires_grad=False)
    batch_seq_len = Variable(torch.LongTensor(np.asarray(seq_lengths)[idx]), requires_grad=False)
    label_seq_lengths = Variable(torch.LongTensor(np.asarray(label_seq_lengths)), requires_grad=False)
    batch_orig = np.asarray(originals)[idx]
    for i,feature in enumerate(batch_features):
        pad_length = max(max_len - len(feature),0)
        buff = np.pad(feature, ((0, pad_length),(0,0)), 'constant', constant_values=( PAD_VALUE))

        batch_features[i] = buff
    batch_features = Variable(torch.FloatTensor(np.stack(batch_features,axis=0)), requires_grad=False)

    return batch_features,batch_label,batch_seq_len,label_seq_lengths,batch_orig
```

```python
   
class CTC(nn.Module):
    def __init__(self,input_size,hidden_size,num_layers,output_size):
        super(CTC,self).__init__()
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.lstm = nn.LSTM(input_size = self.input_size,
                            hidden_size = self.hidden_size,
                            num_layers = self.num_layers,
                            batch_first = True)
        self.linear = nn.Linear(self.hidden_size,output_size)

        
    def forward(self,x):
        output_lstm,(hn,cn) = self.lstm(x)
        output_lin = self.linear(output_lstm)
        output = output_lin.log_softmax(2)
        output = output_lin
        return output

model = CTC(num_features,num_hidden,num_layers,num_classes)
```

```python
optimizer = torch.optim.Adam(lr=initial_learning_rate,params=model.parameters())
ctc_loss = nn.CTCLoss(blank=0)
```

```python
error = []
for i in range(num_epochs):
    x, y_target, seq_len, y_len, seq_orig = batch_gen(features, label_array, seq_lengths, originals)
    y_pred = model.forward(x)
    y_pred = y_pred.permute(1,0,2)
    
    loss = ctc_loss(y_pred, y_target, seq_len, y_len)
    loss.backward()
    torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm = 1)
    optimizer.step()
    if i%100 == 4:
        print(i , ' ',loss.data)
        error.append(loss.data)
    optimizer.zero_grad()
```

```python
def error_calc(a,b):
  right = 0
  for i in range(len(a)):
    for j in range(len(b)):
      if (i == j) and (a[i] == b[j]):
        right = right + 1
  return right/len(b)
```
